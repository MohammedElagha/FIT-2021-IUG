# Data Preprocessing

## Attribute Selection
(according to business objectives)

### fatalities

- drop Country, because iso_o is enough

```
fatalities = fatalities.drop('Country', axis=1)
```

```
fatalities = fatalities.drop('Unnamed: 0', axis=1)
```

### forciby_displaced

- drop CountryOriginName, because CountryOriginCode is enough
- drop CountryAsylumName, because CountryAsylumCode is enough

```
forciby_displaced = forciby_displaced.drop('CountryOriginName', axis=1)
forciby_displaced = forciby_displaced.drop('CountryAsylumName', axis=1)
```

```
forciby_displaced = forciby_displaced.drop('Unnamed: 0', axis=1)
```

### terrain_info

- drop isonum and country, because isocode is enough
- drop rugged_popw, rugged_slope, rugged_lsd and rugged_pc, because rugged is enough
- drop rgdppc_2000, rgdppc_1950_m, rgdppc_1975_m and rgdppc_2000_m, because rgdppc_1950_2000_m is enough
- drop lat and lon, because it descriptive data have no effect on the objective

```
terrain_info = terrain_info.drop('isonum', axis=1)
terrain_info = terrain_info.drop('country', axis=1)
terrain_info = terrain_info.drop('rugged_popw', axis=1)
terrain_info = terrain_info.drop('rugged_slope', axis=1)
terrain_info = terrain_info.drop('rugged_lsd', axis=1)
terrain_info = terrain_info.drop('rugged_pc', axis=1)
terrain_info = terrain_info.drop('rgdppc_2000', axis=1)
terrain_info = terrain_info.drop('rgdppc_1950_m', axis=1)
terrain_info = terrain_info.drop('rgdppc_1975_m', axis=1)
terrain_info = terrain_info.drop('rgdppc_2000_m', axis=1)
terrain_info = terrain_info.drop('lat', axis=1)
terrain_info = terrain_info.drop('lon', axis=1)
```

### geo_cepii

- drop iso2 and cnum, becayse iso3 is enough and equal to used selected country code in previous datasets
- drop country and pays, becayse iso3 is enough
- drop city_en and city_fr
- drop lat and lon, because it descriptive data have no effect on the objective
- drop langoff_2, langoff_3, lang20_1, lang20_2, lang20_3, lang20_4, lang9_1, lang9_2, lang9_3 and lang9_4, because we need only one 'lang' representetive feature
- drop colonizer1, colonizer2, colonizer3 and colonizer4, because has no effect and cannot be represented as one feature
- drop short_colonizer1, short_colonizer2 and short_colonizer3, because has no effect and cannot be represented as one feature

```
geo_cepii = geo_cepii.drop('iso2', axis=1)
geo_cepii = geo_cepii.drop('cnum', axis=1)
geo_cepii = geo_cepii.drop('country', axis=1)
geo_cepii = geo_cepii.drop('pays', axis=1)
geo_cepii = geo_cepii.drop('city_en', axis=1)
geo_cepii = geo_cepii.drop('city_fr', axis=1)
geo_cepii = geo_cepii.drop('lat', axis=1)
geo_cepii = geo_cepii.drop('lon', axis=1)
geo_cepii = geo_cepii.drop('langoff_2', axis=1)
geo_cepii = geo_cepii.drop('langoff_3', axis=1)
geo_cepii = geo_cepii.drop('lang20_1', axis=1)
geo_cepii = geo_cepii.drop('lang20_2', axis=1)
geo_cepii = geo_cepii.drop('lang20_3', axis=1)
geo_cepii = geo_cepii.drop('lang20_4', axis=1)
geo_cepii = geo_cepii.drop('lang9_1', axis=1)
geo_cepii = geo_cepii.drop('lang9_2', axis=1)
geo_cepii = geo_cepii.drop('lang9_3', axis=1)
geo_cepii = geo_cepii.drop('lang9_4', axis=1)
geo_cepii = geo_cepii.drop('colonizer1', axis=1)
geo_cepii = geo_cepii.drop('colonizer2', axis=1)
geo_cepii = geo_cepii.drop('colonizer3', axis=1)
geo_cepii = geo_cepii.drop('colonizer4', axis=1)
geo_cepii = geo_cepii.drop('short_colonizer1', axis=1)
geo_cepii = geo_cepii.drop('short_colonizer2', axis=1)
geo_cepii = geo_cepii.drop('short_colonizer3', axis=1)
```

# Modeling

```
df = SYR_aggr
```

```
PCA_df = df.drop(['REF','IDP','ASY'], axis=1)
PCA_df = df.drop(['date_time'], axis=1)
```

```
pca = PCA(n_components=1)
principalComponents_df = pca.fit_transform(PCA_df)
new_df = pd.DataFrame(data = principalComponents_df, columns = ['component'])
```

```
new_df['REF'] = df['REF']
new_df['IDP'] = df['IDP']
new_df['ASY'] = df['ASY']
new_df['date_time'] = df['date_time']
```

```
import fbprophet
import matplotlib.pyplot as plt
```

```
REF_df = new_df[['REF', 'date_time']]

# Prophet requires columns ds (Date) and y (value)
REF_df = REF_df.rename(columns={'date_time': 'ds', 'REF': 'y'})
REF_df

# Make the prophet model and fit on the data
data_prophet = fbprophet.Prophet(changepoint_prior_scale=20, changepoint_range=0.8, weekly_seasonality=False, daily_seasonality=False)
data_prophet.fit(REF_df)

# Make a future dataframe for 5 years
data_forecast = data_prophet.make_future_dataframe(periods=5, freq='Y')

# Make predictions
data_forecast = data_prophet.predict(data_forecast)

data_prophet.plot(data_forecast, xlabel = 'Date', ylabel = 'REF')
plt.title('REF')
```

```
IDP_df = new_df[['IDP', 'date_time']]

# Prophet requires columns ds (Date) and y (value)
IDP_df = IDP_df.rename(columns={'date_time': 'ds', 'IDP': 'y'})
IDP_df

# Make the prophet model and fit on the data
data_prophet = fbprophet.Prophet(changepoint_prior_scale=20, changepoint_range=0.8, weekly_seasonality=False, daily_seasonality=False)
data_prophet.fit(IDP_df)

# Make a future dataframe for 5 years
data_forecast = data_prophet.make_future_dataframe(periods=5, freq='Y')

# Make predictions
data_forecast = data_prophet.predict(data_forecast)

data_prophet.plot(data_forecast, xlabel = 'Date', ylabel = 'IDP')
plt.title('IDP')
```
